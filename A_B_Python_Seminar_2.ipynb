```python
df_group_A = df.query('con_treat=="control" and page!="new_page"')
df_group_B = df.query('con_treat=="treatment" and page!="old_page"')
```

```python
df_group_A.describe()
```

```python
df_group_B.describe()
```

```python
k1 = df_group_A.converted.sum()
k2 = df_group_B.converted.sum()

n1 = df_group_A.shape[0]
n2 = df_group_B.shape[0]

k1, k2, n1, n2
```

```python
from statsmodels.stats import proportion
chisq, pvalue, table = proportion.proportions_chisquare(np.array([k1, k2]),
np.array([n1, n2]))

print('Results are ','chisq =%.3f, pvalue = %.3f'%(chisq, pvalue))
```

```python
z_score, z_pvalue = proportion.proportions_ztest(np.array([k1, k2]),
                                                 np.array([n1, n2]))
print('Results are ','z_score =%.3f, pvalue = %.3f'%(z_score, z_pvalue))
```
